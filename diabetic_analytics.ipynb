```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading the data from the tsv file
df=pd.read_csv("diabetese.tsv",sep="\t")
df.head()
```

```python
df.info()
```

```python
# What is the average age of diabetic patients (When there is not outliers it is helpful)
mean_age=df["AGE"].mean()
print("Mean age is",mean_age) 
```

```python
# Variance - how the values differ from mean
var_age=df["AGE"].var()
print("Variance in age is",var_age)
```

```python
# Standard Deviation 
std_age=df["AGE"].std()
print("Standard Deviation in age is",std_age)
```

```python
# Box Plotting from age field
plt.figure(figsize=(2,6))
plt.boxplot(df["AGE"],showmeans=True)
plt.grid(color='gray',linestyle='dotted')
plt.xlabel("Age")
plt.ylabel("Years")
plt.show()
```

### We get following info from above box plot
1. 25 percent of people who are diabetic are less than 38 years of age
2. Mean age is around 48
3. 50 percent of people who are diabetic are less than 50 years of age
4. 75 percent of people who are diabetic are less than 58 years of age


```python
# Box Plotting for BMI Field
plt.figure(figsize=(6,2))
plt.boxplot(df["BMI"],vert=False,showmeans=True)
plt.grid(color='gray',linestyle='dotted')
plt.xlabel("Value")
plt.ylabel("BMI")
plt.show()
```

```python
# Box Plotting for BMI Field
plt.figure(figsize=(6,2))
plt.boxplot(df["BP"],vert=False,showmeans=True)
plt.grid(color='gray',linestyle='dotted')
plt.xlabel("Value")
plt.ylabel("BP")
plt.show()
```

```python
# What age of the patient are mostly diabetic 
# For this we create a histogram
df['AGE'].hist(bins=12)                # bins here represents range
plt.title('Age distribution of diabetic patients')
plt.xlabel("Age Range")
plt.ylabel("Count")
plt.show()
# From this we get that most of the patient who are diabetic have an age of around 50
```

```python
df['BMI'].hist(bins=12)                # bins here represents range
plt.title('BMI distribution of diabetic patients')
plt.xlabel("BMI Range")
plt.ylabel("Count")
plt.show()
```

```python
# Scatter plot is used to identify the relationship between attributes of dataset
study_hours=[2,3,1,4,2,5,4,6,3,5]
exam_scores=[70,80,60,85,75,90,88,92,80,95]

plt.scatter(study_hours,exam_scores,color='red',marker='o')
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.title("Scatter Plot : Study Hours vs Exam Scores")
plt.show()

# We see that exam scores is directly proportional to study hours for a student
```

```python
# BMI vs Disease Progression
plt.scatter(df['BMI'],df['Y'],color='red',marker='o')
plt.xlabel("BMI")
plt.ylabel("Disease Progression")
plt.title("Scatter Plot : BMI vs Disease Progression")
plt.show()
```

```python
# S5 vs Disease Progression 
plt.scatter(df['S5'],df['Y'],color='red',marker='o')
plt.xlabel("Blood Type")
plt.ylabel("Disease Progression")
plt.title("Scatter Plot : S5  vs Disease Progression")
plt.show()
```
